```python
import os
import time
import csv
import numpy as np
from scipy import interpolate, integrate, optimize
import matplotlib.pyplot as plt

# ─── CONSTANTS ────────────────────────────────────────────────────────────────
FIXED_INIT    = np.array([1.2e-4, 0.05, 0.03, 11.25, 0.08, 42.9])
TIME_BEFORE22 = np.array([0,2,3,4,5,6,8,10,12,14,16,19,22])
TIME_POINTS   = np.array([22,24,25,27,30,40,50,60,70,80,90,100,120,140,160,180])
ALPHA         = 2.0

# ─── LOAD 25FSIGT DATA ─────────────────────────────────────────────────────────
Glu25   = np.genfromtxt('25FSIGT_Glu.csv', delimiter=',')
FFA25   = np.genfromtxt('25FSIGT_FFA.csv', delimiter=',')
Ins25   = np.genfromtxt('25FSIGT_Ins.csv', delimiter=',')
GIF0_25 = np.genfromtxt('25FSIGT_GIF0.csv', delimiter=',')
InsB22  = np.genfromtxt('Insbefore22_25FSIGT.csv', delimiter=',')
subjects25 = {
    i+1: {
        'tp': TIME_POINTS,
        'glu': Glu25[i],
        'ins': Ins25[i],
        'ffa': FFA25[i],
        'gb': GIF0_25[i,0],
        'ib': GIF0_25[i,1],
        'fb': GIF0_25[i,2]
    }
    for i in range(25)
}

# ─── MODEL COMPONENTS ─────────────────────────────────────────────────────────
def rhs_ffa(params, ins_func, gb, fb, ib):
    si, cx, sg, x2, cf, l2 = params
    def deriv(t, state):
        glu, ffa, xt = state
        return [
            sg*(gb-glu) - si*xt*glu,
            -cf*(ffa-fb) - l2*(xt/x2)**ALPHA/(1+(xt/x2)**ALPHA),
            cx*(max(ins_func(t)-ib,0.0)-xt)
        ]
    return deriv

def compute_xt22(cx, ins_func, ib):
    def dxdt(t, x):
        return cx*(max(ins_func(t)-ib,0.0)-x)
    sol = integrate.solve_ivp(
        dxdt, [0,22], [0.0],
        method='RK45', t_eval=[22],
        rtol=1e-12, atol=1e-12
    )
    return sol.y[0,-1]

# ─── MAIN FUNCTION ────────────────────────────────────────────────────────────
def optimize_and_save(subject_id, date_str='250521'):
    """
    Optimize subject_id (1–25), print per-iteration stats,
    save subj_<id>_<date>.csv under optimization_<date>/,
    and plot the fit.
    """
    if not 1 <= subject_id <= 25:
        raise ValueError("subject_id must be in 1..25")

    # unpack data
    sub = subjects25[subject_id]
    tp, glu_dat, ins_dat, ffa_dat = sub['tp'], sub['glu'], sub['ins'], sub['ffa']
    gb, ib, fb                = sub['gb'], sub['ib'], sub['fb']
    ins_func    = interpolate.interp1d(tp, ins_dat, kind='linear', fill_value='extrapolate')
    insb22_func = interpolate.interp1d(TIME_BEFORE22, InsB22[subject_id-1],
                                       kind='linear', fill_value='extrapolate')

    # counters for printing
    counters = {'nfev': 0, 'last_cost': None, 'nit': 0}

    def cost_fn(x):
        counters['nfev'] += 1
        p = np.abs(x)
        xt22 = compute_xt22(p[1], insb22_func, ib)
        sol = integrate.solve_ivp(
            rhs_ffa(p, ins_func, gb, fb, ib),
            [tp[0], tp[-1]],
            [glu_dat[0], ffa_dat[0], xt22],
            t_eval=tp, method='LSODA',
            rtol=1e-12, atol=1e-12
        ).y
        c1 = np.mean((sol[0]-glu_dat)**2 / glu_dat**2)
        c2 = np.mean((sol[1]-ffa_dat)**2 / ffa_dat**2)
        cost = c1 + c2
        counters['last_cost'] = cost
        return cost

    def iter_cb(xk):
        counters['nit'] += 1
        print(f" Subj {subject_id} — iter {counters['nit']:3d} "
              f"| nfev {counters['nfev']:5d} | cost {counters['last_cost']:.3e}")

    print(f"\n>>> Starting subj {subject_id} (init={FIXED_INIT})")
    t0 = time.perf_counter()
    res = optimize.minimize(
        cost_fn,
        FIXED_INIT,
        method='Nelder-Mead',
        callback=iter_cb,
        options={'maxiter':2000, 'disp':False}
    )
    elapsed = time.perf_counter() - t0

    print(f"→ Done subj {subject_id}: success={res.success}, "
          f"nit={res.nit}, nfev={res.nfev}, cost={res.fun:.3e}, time={elapsed:.2f}s\n")

    # prepare summary
    row = {
        'subject_id': subject_id,
        'date':       date_str,
        'si':         res.x[0],
        'cx':         res.x[1],
        'sg':         res.x[2],
        'x2':         res.x[3],
        'cf':         res.x[4],
        'l2':         res.x[5],
        'cost':       res.fun,
        'nfev':       res.nfev,
        'nit':        res.nit,
        'time_sec':   elapsed,
        'success':    res.success,
        'message':    res.message
    }

    # save CSV
    out_dir = f"optmGFX3Dlocal_{date_str}"
    os.makedirs(out_dir, exist_ok=True)
    csv_path = os.path.join(out_dir, f"subj_{subject_id:02d}_{date_str}.csv")
    with open(csv_path, 'w', newline='') as f:
        w = csv.DictWriter(f, fieldnames=list(row.keys()))
        w.writeheader()
        w.writerow(row)
    print(f"Saved summary → {csv_path}")

    # plot fit
    p_opt = np.abs(res.x)
    xt22 = compute_xt22(p_opt[1], insb22_func, ib)
    sol  = integrate.solve_ivp(
        rhs_ffa(p_opt, ins_func, gb, fb, ib),
        [tp[0], tp[-1]],
        [glu_dat[0], ffa_dat[0], xt22],
        t_eval=tp, method='LSODA',
        rtol=1e-12, atol=1e-12
    ).y

    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.plot(tp, sol[0], '-o', label='Pred')
    plt.plot(tp, glu_dat, 'x', label='Data')
    plt.title(f'Subj {subject_id} – Glucose'); plt.xlabel('Time'); plt.ylabel('Glu'); plt.legend()

    plt.subplot(1,2,2)
    plt.plot(tp, sol[1], '-o', label='Pred')
    plt.plot(tp, ffa_dat, 'x', label='Data')
    plt.title(f'Subj {subject_id} – FFA'); plt.xlabel('Time'); plt.ylabel('FFA'); plt.legend()

    plt.tight_layout()
    plt.show()

    
def optimize_range(start_id, end_id, date_str='250521'):
    """
    Run optimize_and_save for every subject from start_id to end_id (inclusive).
    """
    for sid in range(start_id, end_id+1):
        optimize_and_save(sid, date_str)

# Then to run subjects 3 through 25:
optimize_range(1, 25, date_str='250521')
```
